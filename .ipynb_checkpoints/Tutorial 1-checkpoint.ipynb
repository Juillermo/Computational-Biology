```python
%matplotlib inline

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf
from sklearn import preprocessing
```

```python
data = pd.read_excel('Cell-Cycle-Set.xlsx')
data.info()
```

```python
data.head(1)
```

```python
data.describe()
```

```python
data = data.dropna(how='any')
len(data)
```

```python
fig, axs = plt.subplots(1,2, figsize=(20, 5))
data.plot(ax = axs[0])
data.plot.hist(ax = axs[1], alpha=0.5)
```

```python
labels = (('mean_RNA_G1', 'mean_RNA_S', 'mean_RNA_G2'), ('mean_protein_G1', 'mean_protein_S', 'mean_protein_G2'))
fig, axs = plt.subplots(1,3, figsize=(20, 5))
data.loc[:,['mean_RNA_G1', 'mean_protein_G1']].plot.hist(ax=axs[0])
data.loc[:,['mean_RNA_S', 'mean_protein_S']].plot.hist(ax=axs[1])
data.loc[:,['mean_RNA_G2', 'mean_protein_G2']].plot.hist(ax=axs[2])
```

```python
data = data.reset_index()
data.corr()
```

```python
results1 = smf.ols('mean_protein_G1 ~ mean_RNA_G1', data=data).fit()
results2 = smf.ols('mean_protein_S ~ mean_RNA_S', data=data).fit()
results3 = smf.ols('mean_protein_G2 ~ mean_RNA_G2', data=data).fit()

fig, axs = plt.subplots(1,3, figsize=(20, 5))
data.plot.scatter(x='mean_RNA_G1', y='mean_protein_G1', ax=axs[0], color='Red');
axs[0].plot(data['mean_RNA_G1'], results1.fittedvalues)
axs[0].set_title('Rsquared: '+str(results1.rsquared));
data.plot.scatter(x='mean_RNA_S',  y='mean_protein_S',  ax=axs[1], color='Yellow');
axs[1].plot(data['mean_RNA_S'], results2.fittedvalues)
axs[1].set_title('Rsquared: '+str(results2.rsquared));
data.plot.scatter(x='mean_RNA_G2', y='mean_protein_G2', ax=axs[2], color='Red');
axs[2].plot(data['mean_RNA_G2'], results3.fittedvalues)
axs[2].set_title('Rsquared: '+str(results3.rsquared))
```

There's a **not-neat correlation** (R^2 ~ 0.3) because of different factors, such as:
- The imprecision of measuring tools
- The differences in proteins lifespan
- Complex interaction between proteins

```python
def scatter(parent_scatter, son_scatter, labels):
    
    results = []
    for i, label in enumerate(labels[0]):
        results.append(smf.ols( label + ' ~ ' + labels[0][i], data=son_scatter).fit())

    fig, axs = plt.subplots(1,3, figsize=(20, 5))

    for i, result in enumerate(results):
        parent_scatter.plot.scatter(x=labels[0][i], y=labels[1][i], ax=axs[i], color='Red', label='All', alpha=0.3);
        son_scatter.plot.scatter(x=labels[0][i], y=labels[1][i], ax=axs[i], color='Blue', label='Sub-guys');
        axs[i].set_title('Rsquared: '+str(result.rsquared));
```


### 1. Find all guys that contain 'cell cycle in GOBP

```python
cell_cycle_guys = data[data['GOBP'].str.contains('cell cycle')]
scatter(data, cell_cycle_guys, labels)
```

### 2. Find the guys that contain 'ribosome' in their GOCC

```python
ribosome_guys = data[data['GOCC'].str.contains('ribosome')]
scatter(data, ribosome_guys, labels)
```

The *ribosomes* dots form two distinct clusters prevalent in the three stages of the cell cycle.

### 4. Calculate the change in mRNA/protein level across the cell cycle

 by taking the difference at each stage (G1-S, S-G2, G2-G1), and standardize the differences by mean-centering and variance scaling. Repeat tasks 1 and 2 by plotting the changes in levels with GOBP/GOCC labelling. What do we notice about changes in the cell cycle? Is there any apparent clustering of GO terms?

```python
data.head(1)
```

```python
G1_S = data.mean_RNA_G1 - data.mean_RNA_S
S_G2 = data.mean_RNA_S - data.mean_RNA_G2
G2_G1 = data.mean_RNA_G2 - data.mean_RNA_G1

G1_Spro = data.mean_protein_G1 - data.mean_protein_S
S_G2pro = data.mean_protein_S - data.mean_protein_G2
G2_G1pro = data.mean_protein_G2 - data.mean_protein_G1

diff_labels = (('G1_S', 'S_G2', 'G2_G1'), ('G1_Spro', 'S_G2pro', 'G2_G1pro'))
diff = pd.DataFrame({'G1_S': G1_S, 'S_G2': S_G2, 'G2_G1': G2_G1, 'G1_Spro': G1_Spro, 'S_G2pro': S_G2pro, 'G2_G1pro': G2_G1pro})

scaler = preprocessing.StandardScaler()
np_scaled = scaler.fit_transform(diff)
diff_norm = pd.DataFrame(np_scaled, columns= [item for sublist in diff_labels for item in sublist])

diff_norm[['GOBP', 'GOMF', 'GOCC']] = data.iloc[:, 8:11]
```

```python
results1 = smf.ols('G1_Spro ~ G1_S', data=diff_norm).fit()
results2 = smf.ols('S_G2pro ~ S_G2', data=diff_norm).fit()
results3 = smf.ols('G2_G1pro ~ G2_G1', data=diff_norm).fit()

fig, axs = plt.subplots(1,3, figsize=(20, 5))
diff_norm.plot.scatter(x='G1_S', y='G1_Spro', ax=axs[0], color='Red');
axs[0].plot(diff_norm['G1_S'], results1.fittedvalues)
axs[0].set_title('Rsquared: '+str(results1.rsquared));
diff_norm.plot.scatter(x='S_G2',  y='S_G2pro',  ax=axs[1], color='Yellow');
axs[1].plot(diff_norm['S_G2'], results2.fittedvalues)
axs[1].set_title('Rsquared: '+str(results2.rsquared));
diff_norm.plot.scatter(x='G2_G1', y='G2_G1pro', ax=axs[2], color='Red');
axs[2].plot(diff_norm['G2_G1'], results3.fittedvalues)
axs[2].set_title('Rsquared: '+str(results3.rsquared))
```

```python
cell_cycle_guys2 = diff_norm[diff_norm['GOBP'].str.contains('cell cycle')]
scatter(diff_norm, cell_cycle_guys2, diff_labels)
```

It's quite significative that the RNA that contain "cell cycle" are precisely the ones with most variance with respect to the proteins, particularly between the steps G2 and G1.

```python
ribosome_guys2 = diff_norm[diff_norm['GOCC'].str.contains('ribosome')]
scatter(diff_norm, ribosome_guys2, diff_labels)
```
